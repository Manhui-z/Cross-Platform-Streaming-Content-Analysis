#Rank the recommendations by both IMDB rating and ReelGood rating,

```python
def recommend_based_on_input(input_text, reelgood_data):
    # Convert input text to lowercase for case-insensitive matching
    input_text = input_text.lower()


    filtered_data = reelgood_data[reelgood_data['Title'].str.lower().str.contains(input_text) |
                                  reelgood_data['Genre'].str.lower().str.contains(input_text)]


    filtered_data = filtered_data.sort_values(by=['IMDB', 'ReelGood'], ascending=[False, False])

    filtered_data = filtered_data.drop_duplicates(subset='Title')


    filtered_data = filtered_data.dropna(subset=["What it's about", 'Where to Watch'])


    recommendations = filtered_data.head(5)[['Title', 'IMDB', 'ReelGood',"What it's about", 'Where to Watch']]

    return recommendations


user_input = input("Please provide a genre of movies or shows you like: ")

recommendations = recommend_based_on_input(user_input, reelgood_data)

if recommendations.empty:
    print("Sorry, no recommendations found based on your input.")
else:
    print("Recommendations based on your input:")
    print(recommendations)
```
